### Практическое занятие по Pandas

```python
import pandas as pd
```

Используемый `data set` (набор данных) для аналитики - `titanic.csv`.

```python
data_frame = pd.read_csv("https://raw.githubusercontent.com/jorisvandenbossche/pandas-tutorial/master/data/titanic.csv")
```

Выведем наши данные. Лучше выводить без метода `print()`, так как функция `print()` изменяет внешний вид таблицы.

```python
data_frame
```

Выведем на экран общую информацию, чтобы понимать из чего состоит наш `data frame` при помощи метода `.info()`.

```python
data_frame.info()
```

```python
data_frame.sort_values(["Age"])[:15]
```

```python
data_frame.sort_values(["Age", "Name"], ascending = [False, True])[ : 15]
```

Получение полной статистики о `data frame`-е.

```python
data_frame.describe()
```

Выведем корреляцию данных по парно.

```python
data_frame.corr()
```

Визуализируем данные.

```python
data_frame["Survived"].plot(kind = "hist", bins = 3)
```

```python
data_frame["Age"].plot(kind = "hist", bins = 70)
```

```python
data_frame["Pclass"].plot(kind = "hist", bins = 25)
```

```python
data_frame["Age"].plot(kind = "kde", xlim = [0.0001, 85])
```

```python
data_frame.groupby("Sex")["Age"].plot(kind = "kde", xlim = [0, 85], legend = True) # legend - подписи, xlim - рассматриваемая часть. 
```
